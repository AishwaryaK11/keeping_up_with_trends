```python
from autoscraper import AutoScraper

url = 'https://www.amazon.in/s?k=womens+cloths+trending&crid=2O5SU3W64608&sprefix=womens+cloths+trending%2Caps%2C306&ref=nb_sb_noss_2'

# We can add one or multiple candidates here.
# You can also put urls here to retrieve urls.
wanted_list = ["top Trending Crop top for Women/Girls","₹379","47"]

scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar(url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_girq':'Title','rule_gvy3':'Price','rule_bp1h':'Rating'})
scraper.keep_rules(['rule_girq','rule_gvy3','rule_bp1h'])
scraper.save('amazn-search')
```
